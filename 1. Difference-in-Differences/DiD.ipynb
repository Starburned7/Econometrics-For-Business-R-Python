```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("njmin3.csv")
```

```python
missing_values = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
missing_values.fit(df[['fte', 'demp']])
```

```python
df[['fte', 'demp']]= missing_values.transform(df[['fte', 'demp']])
```
